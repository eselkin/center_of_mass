```python
import cv2
import matplotlib.pyplot as plt
import sklearn
from sklearn import cluster
import numpy as np
%matplotlib inline
```

```python
cap = cv2.VideoCapture('/Volumes/MOUSE/Analog_20180105_1246_mpeg2video.mpg')
cv2.__version__
```

```python
# Not necessary after the gaussian blurs
def separate(split_, fgmask_):
    split_[fgmask_<=200] = [0,0,0]
    lower_white = np.array([0,10,10], dtype = "uint16")
    upper_white = np.array([255,255,255], dtype = "uint16")
    white_mask = cv2.inRange(split_, lower_white, upper_white)
    split_[white_mask == 255] = [255,255,255]
    split_[white_mask == 0] = [0,0,0]
    return split_

def blur(split_):
    return (cv2.GaussianBlur(split_, (3,3), 0) - cv2.GaussianBlur(split_, (5,5), 1) + cv2.GaussianBlur(split_, (7,7), 0))

def findCentroid(fgmask, size = 600):
    fgmask[fgmask == 255] = 100
    _, c, _ = cv2.findContours(fgmask, 1, 2)
    centroid = None
    if (len(c) > 0):
        cnt = max(c, key = cv2.contourArea)
        if (cv2.contourArea(cnt) > size):
            M_0 = cv2.moments(cnt)
            cx = int(M_0['m10']/M_0['m00'])
            cy = int(M_0['m01']/M_0['m00'])
            fgmask[cy-5:cy+5,cx-5:cx+5] = 255
            centroid = np.asarray([cy, cx]) # row, col
    return fgmask, centroid
```

```python
# Standard MOG, but shows fewer mouse edges
fgbg_0 = cv2.bgsegm.createBackgroundSubtractorMOG(backgroundRatio=0.5, history=30)
fgbg_1 = cv2.bgsegm.createBackgroundSubtractorMOG(backgroundRatio=0.5, history=30)
fgbg_2 = cv2.bgsegm.createBackgroundSubtractorMOG(backgroundRatio=0.5, history=30)
fgbg_3 = cv2.bgsegm.createBackgroundSubtractorMOG(backgroundRatio=0.5, history=30)

# CNT (count) gives better edges but leaves trailing histories of fg activity
# fgbg_0 = cv2.bgsegm.createBackgroundSubtractorCNT()
# fgbg_1 = cv2.bgsegm.createBackgroundSubtractorCNT()
# fgbg_2 = cv2.bgsegm.createBackgroundSubtractorCNT()
# fgbg_3 = cv2.bgsegm.createBackgroundSubtractorCNT()

# For morphological opening later
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
centroids_0 = []
centroids_1 = []
centroids_2 = []
centroids_3 = []
writer = cv2.VideoWriter("output.avi", cv2.VideoWriter_fourcc(*"MJPG"), 30,(325,150))
i = 0
while(1):
    i += 1
    ret, frame = cap.read()
    # Defining major edges but also blurring the artifacts of the video away (lines error noise etc)
    #frame = cv2.medianBlur(frame, 3) - cv2.GaussianBlur(frame, (5,5), 0) + cv2.GaussianBlur(frame, (7,7), 0)

    # Separate the quad into separate sections, removing the tops of the cages from the view ~90pixels
    cage_top_height = 90
    margin_l = 20
    margin_r_l = 5
    margin_r = 15
    split_0 = np.copy(frame[cage_top_height:480/2, margin_l:(720/2-margin_r), :])
    split_0 = blur(split_0)
    split_1 = np.copy(frame[cage_top_height:480/2, (720/2-margin_r_l):720-margin_r, :])
    split_1 = blur(split_1)
    split_2 = np.copy(frame[480/2+cage_top_height:480, margin_l:(720/2-margin_r), :])
    split_2 = blur(split_2)
    split_3 = np.copy(frame[480/2+cage_top_height:480, (720/2-margin_r_l):720-margin_r, :])
    split_3 = blur(split_3)
    
    # Apply the background subtraction masks to each section
    # Each section has its own subtractor, since they each develop a background model
    fgmask_0 = fgbg_0.apply(split_0)
    fgmask_0 = cv2.morphologyEx(fgmask_0, cv2.MORPH_OPEN, kernel).reshape(list(fgmask_0.shape[:2])+[1])
    fgmask_0, center_0 = findCentroid(fgmask_0)
    fgmask_1 = fgbg_1.apply(split_1)
    fgmask_1 = cv2.morphologyEx(fgmask_1, cv2.MORPH_OPEN, kernel)
    fgmask_1, center_1 = findCentroid(fgmask_1)
    fgmask_2 = fgbg_2.apply(split_2)
    fgmask_2 = cv2.morphologyEx(fgmask_2, cv2.MORPH_OPEN, kernel)
    fgmask_2, center_2 = findCentroid(fgmask_2)
    fgmask_3 = fgbg_3.apply(split_3)
    fgmask_3 = cv2.morphologyEx(fgmask_3, cv2.MORPH_OPEN, kernel)
    fgmask_3, center_3 = findCentroid(fgmask_3)
    centroids_0.append(center_0)
    centroids_1.append(center_1)
    centroids_2.append(center_2)
    centroids_3.append(center_3)
    cv2.imshow('split0',fgmask_0)
    deep = np.dstack([fgmask_0, fgmask_0, fgmask_0])
    writer.write(deep)
    cv2.imshow('split1',fgmask_1)
    cv2.imshow('split2',fgmask_2)
    cv2.imshow('split3',fgmask_3)
    cv2.imshow("whole", frame)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
    if i == 10000:
        writer.relase()
        break;
cap.release()
cv2.destroyAllWindows()
```
